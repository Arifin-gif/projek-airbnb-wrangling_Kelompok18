```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import warnings

# Abaikan warning agar output bersih
warnings.filterwarnings('ignore')

print("‚úÖ Library berhasil di-load.")

# --- FUNGSI PEMBANTU (PERMANEN) ---
def clean_price_column(df, column_name='price'):
    """Membersihkan kolom harga dari simbol '$' dan ',' lalu konversi ke float."""
    if column_name in df.columns:
        df[f'{column_name}_cleaned'] = (
            df[column_name].astype(str)
            .str.replace('$', '', regex=False)
            .str.replace(',', '', regex=False)
            .astype(float)
        )
    return df
```

```python
#  MEMUAT & MEMBERSIHKAN DATA
print("--- 1. MEMULAI PROSES LOAD & CLEAN ---")

files_map = {
    'Jersey City': {'list': 'Jersey City_listings.csv', 'cal': 'Jersey City_calendar.csv'},
    'Albany':      {'list': 'Albany_listings.csv',      'cal': 'Albany_calendar.csv'},
    'Twin Cities': {'list': 'Twin Cities_listings.csv', 'cal': 'Twin Cities_calendar.csv'}
}

processed_data = {} # Tempat menyimpan data bersih

for city, files in files_map.items():
    print(f"\nProcessing {city}...")

    # A. PROSES LISTINGS
    if os.path.exists(files['list']):
        df_list = pd.read_csv(files['list'])
        # 1. Clean Price
        df_list = clean_price_column(df_list, 'price')
        # 2. Clean Host Rate
        if 'host_response_rate' in df_list.columns:
            df_list['host_response_rate_cleaned'] = df_list['host_response_rate'].str.replace('%', '', regex=False).astype(float) / 100
            df_list['host_response_rate_cleaned'] = df_list['host_response_rate_cleaned'].fillna(0)
        # 3. Rename ID
        if 'id' in df_list.columns:
            df_list = df_list.rename(columns={'id': 'listing_id'})

        df_list['city'] = city
        processed_data[f"{city} Listings"] = df_list
        print(f"  ‚úÖ Listings Loaded & Cleaned ({df_list.shape[0]} baris)")
    else:
        print(f"  ‚ùå Listings File Missing: {files['list']}")

    # B. PROSES CALENDAR
    if os.path.exists(files['cal']):
        df_cal = pd.read_csv(files['cal'])
        # 1. Clean Price
        df_cal = clean_price_column(df_cal, 'price')
        # 2. Clean Date & Availability
        if 'date' in df_cal.columns:
            df_cal['date'] = pd.to_datetime(df_cal['date'])
        if 'available' in df_cal.columns:
            df_cal['is_available'] = df_cal['available'].map({'t': 1, 'f': 0})

        df_cal['city'] = city
        processed_data[f"{city} Calendar"] = df_cal
        print(f"  ‚úÖ Calendar Loaded & Cleaned ({df_cal.shape[0]} baris)")
    else:
        print(f"  ‚ùå Calendar File Missing: {files['cal']}")
```

```python
# INTEGRASI & FINALISASI DATA
print("\n--- 2. INTEGRASI & FINALISASI DATA ---")

cols_to_keep = [
    'listing_id', 'price_cleaned', 'host_response_rate_cleaned',
    'neighbourhood_cleansed', 'room_type', 'accommodates', 'review_scores_rating'
]

final_dfs = []

# Loop Integrasi
for city in files_map.keys():
    cal_key = f"{city} Calendar"
    list_key = f"{city} Listings"

    if cal_key in processed_data and list_key in processed_data:
        available_cols = [c for c in cols_to_keep if c in processed_data[list_key].columns]
        merged_city = pd.merge(
            processed_data[cal_key],
            processed_data[list_key][available_cols],
            on='listing_id', how='left'
        )
        merged_city['city_name'] = city
        final_dfs.append(merged_city)
        print(f"‚úÖ Data {city} siap.")

# Gabung Semua Kota
if final_dfs:
    final_master_df = pd.concat(final_dfs, ignore_index=True)

    # Rename Kolom (Sesuai kode permanen)
    rename_map = {
        'city_name': 'city',
        'listing_id': 'listing_id',
        'date': 'date',
        'is_available': 'availability_status',
        'price_cleaned_y': 'static_price_usd',
        'price_cleaned_x': 'daily_price_nan',
        'price_cleaned': 'daily_price_nan',
        'neighbourhood_cleansed': 'neighborhood',
        'room_type': 'room_type',
        'accommodates': 'accommodates',
        'review_scores_rating': 'review_rating',
        'host_response_rate_cleaned': 'host_response_rate_decimal'
    }

    available_rename = [c for c in rename_map.keys() if c in final_master_df.columns]
    final_df = final_master_df[available_rename].rename(columns=rename_map)

    # Imputasi Rating
    if 'review_rating' in final_df.columns:
        median_rate = final_df['review_rating'].median()
        final_df['review_rating'] = final_df['review_rating'].fillna(median_rate)

    # Export CSV
    output_filename = 'Project_Final_Data_Gabungan.csv'
    final_df.to_csv(output_filename, index=False)
    print(f"\n‚úÖ File Final Diekspor: {output_filename}")
    print(f"‚úÖ Total Data: {len(final_df):,} baris")
else:
    print("‚ùå Error: Tidak ada data untuk diproses.")
```

```python
# --- MEMBUAT DASHBOARD UTAMA (USER VERSION) ---
if 'final_df' in locals() and not final_df.empty:
    print("\n--- 3. MEMBUAT DASHBOARD UTAMA ---")

    # Filter harga outlier visual (< $500)
    viz_data = final_df[final_df['static_price_usd'] < 500]

    sns.set_theme(style="whitegrid")
    fig, axes = plt.subplots(2, 2, figsize=(18, 14))
    fig.suptitle('Analisis Pasar Airbnb: Jersey City vs Albany vs Twin Cities', fontsize=20, fontweight='bold')

    # Grafik 1: Harga Rata-rata
    sns.barplot(ax=axes[0, 0], x='city', y='static_price_usd', data=viz_data, estimator=np.mean, errorbar=None, palette='viridis')
    axes[0, 0].set_title('Rata-Rata Harga per Malam (USD)')
    axes[0, 0].bar_label(axes[0, 0].containers[0], fmt='$%.0f', padding=3)

    # Grafik 2: Tipe Kamar
    sns.countplot(ax=axes[0, 1], x='city', hue='room_type', data=viz_data, palette='pastel')
    axes[0, 1].set_title('Jumlah Properti berdasarkan Tipe Kamar')

    # Grafik 3: Heatmap Ketersediaan
    viz_data['day_name'] = viz_data['date'].dt.day_name()
    days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    heatmap_data = viz_data.pivot_table(index='day_name', columns='city', values='availability_status', aggfunc='mean') * 100
    heatmap_data = heatmap_data.reindex(days_order)

    sns.heatmap(heatmap_data, ax=axes[1, 0], annot=True, fmt=".1f", cmap="YlGnBu", cbar_kws={'label': 'Ketersediaan (%)'})
    axes[1, 0].set_title('Heatmap Ketersediaan Kamar (%) per Hari')

    # Grafik 4: Boxplot Harga
    sns.boxplot(ax=axes[1, 1], x='city', y='static_price_usd', data=viz_data, palette='Set2')
    axes[1, 1].set_title('Distribusi Harga (Filter < $500)')

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.savefig('Dashboard_Analisis_Airbnb.png', dpi=300)
    plt.show()
```

```python
# BAGIAN 2: ANALISIS LANJUTAN

if 'final_df' in locals() and not final_df.empty:
    print("üöÄ MEMULAI ANALISIS LANJUTAN\n")

    # 1. Cleaning Lebih Ketat
    df_adv = final_df.copy()
    init_rows = len(df_adv)
    df_adv = df_adv.drop_duplicates()
    df_adv = df_adv[df_adv['static_price_usd'] > 0] # Hapus harga 0 atau negatif
    print(f"‚úÖ Advanced Cleaning: {init_rows - len(df_adv)} baris duplikat/invalid dihapus.")

    # 2. Feature Engineering (Waktu)
    df_adv['month'] = df_adv['date'].dt.month_name()
    df_adv['month_num'] = df_adv['date'].dt.month

    # 3. Visualisasi Seasonality (Tren Bulanan)
    print("\nüìä [Grafik 5] Analisis Tren Harga Musiman")
    seasonal_price = df_adv.groupby(['city', 'month_num', 'month'])['static_price_usd'].mean().reset_index()
    seasonal_price = seasonal_price.sort_values('month_num')

    plt.figure(figsize=(14, 6))
    sns.lineplot(data=seasonal_price, x='month', y='static_price_usd', hue='city', marker='o', linewidth=2.5)
    plt.title('Tren Harga Rata-Rata per Bulan (Seasonality)', fontsize=16, fontweight='bold')
    plt.ylabel('Rata-Rata Harga ($)')
    plt.xlabel('Bulan')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
    print("üí° Insight: Grafik ini menunjukkan kapan waktu termahal (High Season) dan termurah (Low Season).")
```

```python
if 'df_adv' in locals():
    # 1. Analisis Lokasi (Neighborhood)
    print("\nüìä [Grafik 6] Top 5 Lokasi (Neighborhood) Termahal")
    cities_list = df_adv['city'].unique()

    fig, axes = plt.subplots(1, 3, figsize=(20, 6), sharey=False)
    fig.suptitle('Top 5 Neighborhood Termahal (Median Price)', fontsize=18, fontweight='bold')

    for i, city in enumerate(cities_list):
        city_data = df_adv[df_adv['city'] == city]
        # Cek apakah kolom neighborhood ada isinya
        if 'neighborhood' in city_data.columns and not city_data['neighborhood'].isnull().all():
            top_hoods = city_data.groupby('neighborhood')['static_price_usd'].median().sort_values(ascending=False).head(5).reset_index()
            sns.barplot(ax=axes[i], data=top_hoods, x='static_price_usd', y='neighborhood', palette='magma')
            axes[i].set_title(city, fontsize=14)
            axes[i].set_xlabel('Median Harga ($)')
            axes[i].set_ylabel('')
        else:
            axes[i].text(0.5, 0.5, 'Data Lokasi Tidak Tersedia', ha='center')
            axes[i].set_title(city)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

    # 2. Heatmap Korelasi
    print("\nüìä [Grafik 7] Heatmap Korelasi Statistik")
    plt.figure(figsize=(8, 6))
    corr_cols = ['static_price_usd', 'review_rating', 'accommodates', 'host_response_rate_decimal', 'availability_status']
    valid_corr = [c for c in corr_cols if c in df_adv.columns]

    if len(valid_corr) > 1:
        sns.heatmap(df_adv[valid_corr].corr(), annot=True, cmap='coolwarm', fmt=".2f")
        plt.title('Korelasi Antar Variabel')
        plt.show()
```

```python
if 'df_adv' in locals():
    print("\nüìä [Grafik 8] Estimasi Pendapatan Tahunan")

    # Asumsi: Jika availability_status = 0, berarti terjual (booked)
    # Revenue = Harga x Jumlah Hari Booked

    revenue_df = df_adv.groupby(['city', 'listing_id']).agg({
        'availability_status': lambda x: (x == 0).sum(), # Hitung hari booked
        'static_price_usd': 'mean'
    }).rename(columns={'availability_status': 'days_booked'}).reset_index()

    revenue_df['est_revenue'] = revenue_df['days_booked'] * revenue_df['static_price_usd']

    plt.figure(figsize=(10, 6))
    sns.barplot(data=revenue_df, x='city', y='est_revenue', estimator=np.mean, errorbar=None, palette='Greens_d')
    plt.title('Estimasi Pendapatan Tahunan Rata-Rata per Listing ($)', fontsize=16)
    plt.ylabel('Revenue ($)')

    ax = plt.gca()
    for container in ax.containers:
        ax.bar_label(container, fmt='$%.0f', padding=3)

    plt.show()
    print("üí° Insight: Angka ini adalah estimasi kasar pendapatan host per tahun berdasarkan kalender.")
    print("\n‚úÖ SELESAI. Seluruh analisis (Dasar + Lanjutan) telah dijalankan.")
```

```python
import scipy.stats as stats

# BAGIAN 3: ANALISIS MENDALAM
# Gunakan bagian ini untuk menjawab pertanyaan statistik & validitas data

if 'df_adv' in locals() and not df_adv.empty:
    print("\n" + "="*60)
    print("üéì BAGIAN 3: STATISTICAL DEEP DIVE (DOSEN Q&A PREPARATION)")
    print("="*60)

    # 1. UJI HIPOTESIS (ANOVA TEST)
    # "Apakah perbedaan harga antar kota ini signifikan secara statistik?"
    print("\nüß™ [1] Uji Statistik (ANOVA Test) pada Harga")

    # Mengambil sampel harga per kota (untuk menghindari bias data terlalu besar)
    # Kita ambil list harga bersih (tanpa outlier ekstrem untuk uji statistik)
    stats_data = df_adv[df_adv['static_price_usd'] < 500]

    prices_jersey = stats_data[stats_data['city'] == 'Jersey City']['static_price_usd']
    prices_albany = stats_data[stats_data['city'] == 'Albany']['static_price_usd']
    prices_twin = stats_data[stats_data['city'] == 'Twin Cities']['static_price_usd']

    # Melakukan One-Way ANOVA
    f_stat, p_value = stats.f_oneway(prices_jersey, prices_albany, prices_twin)

    print(f"   Hasil F-Statistic : {f_stat:.2f}")
    print(f"   P-Value           : {p_value:.4f}")

    if p_value < 0.05:
        print("   üëâ KESIMPULAN: P-Value < 0.05. Perbedaan harga antar kota adalah SIGNIFIKAN (Bukan kebetulan).")
    else:
        print("   üëâ KESIMPULAN: Perbedaan harga TIDAK signifikan.")


    # 2. ANALISIS DISTRIBUSI HARGA (SKEWNESS)
    # "Apakah data harganya terdistribusi normal?"
    print("\nüìä [2] Analisis Distribusi Data (Skewness)")

    plt.figure(figsize=(12, 5))

    # Plot Histogram dengan KDE (Kernel Density Estimate)
    sns.histplot(data=stats_data, x='static_price_usd', hue='city', kde=True, element="step", palette='viridis', alpha=0.3)

    plt.title('Distribusi Harga Properti (Filtered < $500)', fontsize=14)
    plt.xlabel('Harga ($)')
    plt.ylabel('Frekuensi')
    plt.show()

    skew_val = stats_data['static_price_usd'].skew()
    print(f"   Nilai Skewness: {skew_val:.2f}")
    if skew_val > 1:
        print("   üëâ Insight: Data 'Positively Skewed' (Miring ke Kanan). Artinya lebih banyak properti murah, tapi ada sedikit properti yang sangat mahal.")
    elif skew_val < -1:
        print("   üëâ Insight: Data 'Negatively Skewed'.")
    else:
        print("   üëâ Insight: Data terdistribusi cukup normal.")


    # 3. FAKTOR PENENTU HARGA (CORRELATION RANKING)
    #  "Faktor apa yang paling mempengaruhi harga sewa?"
    print("\nww [3] Faktor Korelasi Terhadap Harga")

    # Hitung korelasi semua variabel numerik terhadap 'static_price_usd'
    numeric_df = df_adv.select_dtypes(include=[np.number])
    correlation_prices = numeric_df.corr()['static_price_usd'].sort_values(ascending=False)

    # Drop harga itu sendiri
    correlation_prices = correlation_prices.drop('static_price_usd')

    print("   Variabel yang paling berkorelasi dengan Harga:")
    print(correlation_prices.to_string())

    print("\n   üëâ Insight:")
    if 'accommodates' in correlation_prices and correlation_prices['accommodates'] > 0.3:
        print("   - Kapasitas (Accommodates) memiliki pengaruh positif terkuat terhadap harga.")
    print("   - Review Rating seringkali memiliki korelasi kecil karena properti murah pun bisa punya rating bagus.")

else:
    print("‚ùå Data 'df_adv' tidak ditemukan. Pastikan sel kode sebelumnya dijalankan.")
```
